#### In the research department of a hospital, a **study is being conducted to determine whether glucose tolerance in healthy patients tends to decrease as a person's age increases**, meaning it takes longer to disappear from our bodies as we age. The glucose tolerance test, also known as the **"oral glucose tolerance test"**, measures the body's response to sugar (glucose). It can be used as a **screening test for type 2 diabetes**.

#### To obtain the study's conclusions, **blood glucose measurement was performed at the time of intake (baseline level) of 100 grams of glucose**, in the form of an oral preparation, **and 60 minutes after intake, on two samples of randomly selected healthy patients divided by age: young (<30 years) and adults (>30 years)**. We will consider these samples independent.

#### The data has a total of 65 records, which are recorded in an Excel file named `'muestra_glucosa.xlsx'`.

#### Data Description
| Variable               | Description                                                                        | Possible values / example      |
| ---------------------- | ---------------------------------------------------------------------------------- | ------------------------------ |
| **Control Group**      | Indicates the patient's age range.                                                 | 1 = Young (<30 years) / 2 = Adult (>30 years) |
| **Baseline Glucose Level** | Blood glucose level (mg/dL) at the time of ingesting 100 g of glucose.              | 89, 94, 102                    |
| **60-min Glucose Level**   | Blood glucose level (mg/dL) measured 60 minutes after intake.                      | 119, 134, 145                  |

#1. Importing libraries.

```python
# coding: utf-8
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as ss
from scipy.stats import norm
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

### Load the DataFrame.

```python
path='glucose_sample.xlsx'
df=pd.read_excel(
    path,
    names=['group', 'basal_glucose', 'glucose_60min'],
    header=0 )
df
```

# Part I
## Descriptive Analysis of Glucose Levels


### a) Centralization and dispersion measures by control groups.

```python
# Create variables to group by age
g1 = df[df['group'] == 1]
g2 = df[df['group'] == 2]

print(f"Total records in Control Group 1: {len(g1)}\nTotal records in Control Group 2: {len(g2)}")
```

```python
# Create variables for Baseline Glucose by group
# Variable for Baseline Glucose of Group 1
g1_before = g1['basal_glucose']

# Variable for Baseline Glucose of Group 2
g2_before = g2['basal_glucose']
```

```python
# Central tendency measures at baseline 'G1'
g1_mean = round(g1_before.mean(), 2)
g1_median = g1_before.median()
g1_mode = g1_before.mode().tolist()

# Dispersion measures at baseline 'G1'
g1_ran = np.ptp(g1_before)
g1_var = round(g1_before.var(), 2)
g1_std = round(g1_before.std(), 2)
g1_iqr = g1_before.quantile(0.75) - g1_before.quantile(0.25)
```

```python
# Central tendency measures at baseline 'G2'
g2_mean = round(g2_before.mean(), 2)
g2_median = g2_before.median()
g2_mode = g2_before.mode().tolist()

# Dispersion measures 'G2'
g2_ran = np.ptp(g2_before)
g2_var = round(g2_before.var(), 2)
g2_std = round(g2_before.std(), 2)
g2_iqr = g2_before.quantile(0.75) - g2_before.quantile(0.25)
```

```python
# Show results
measures = pd.DataFrame({
    'mean': [g1_mean, g2_mean],
    'median': [g1_median, g2_median],
    'mode': [g1_mode, g2_mode],
    'range': [g1_ran, g2_ran],
    'variance': [g1_var, g2_var],
    'standard_deviation': [g1_std, g2_std],
    'iqr': [g1_iqr, g2_iqr]
}, index=['group_1', 'group_2'])

print("Results of central tendency and dispersion measures of 'Baseline Glucose' by control groups:\n")
display(measures)
print('Glucose levels are expressed in mg/dL.')
```

Los **adultos (g2) tienen en promedio la glucosa basal más alta** que los jóvenes (g1). La **distribución de la media en los adultos (g2) es más simétrica** que la de los jóvenes (g1). El **grupo de jóvenes (g1) es multimodal con 4 modas**, y el de adultos (g2) no. El **rango es mayor en los jóvenes (g1)**. La **varianza, desviación estándar y el IQR son menores en los adultos (g2)**.

#### ¿Es la media representativa para cada grupo?   

Para evaluar si la media es representativa de los datos, se calcula el **coeficiente de variación (CV)**, que expresa la dispersión relativa como porcentaje de la media.

- Si CV ≤ 30% → la media es representativa (baja dispersión relativa)
- Si CV > 30% → la media puede no ser representativa (alta dispersión relativa)

```python
# Hist. de Glucosa Basal en ambos grupos
# g1_before y g2_before
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Grupo 1
g1_plot = g1_before.plot(kind='hist', ax=ax1, color='#E4485C', edgecolor='white', density=True)
g1_plot.set_xlabel('Nivel de Glucosa(mg/dL) ',fontsize=10)
g1_plot.set_ylabel('Frecuencia', fontsize=10)
g1_plot.set_title('Jóvenes (Grupo 1)', fontsize=12, fontweight='bold')
ax1.grid(True, alpha=0.3)


# Grupo 2
g2_plot = g2_before.plot(kind='hist', ax=ax2, color='#065B62', edgecolor='white', density=True)
g2_plot.set_xlabel('Nivel de Glucosa(mg/dL) ', fontsize=10)
g2_plot.set_ylabel('Frecuencia', fontsize=10)
g2_plot.set_title('Adultos (Grupo 2)', fontsize=12, fontweight='bold')
ax2.grid(True, alpha=0.3)


fig.suptitle('Niveles de Glucosa Basal', fontsize=14, fontweight='bold')
fig.tight_layout(pad=1.2)
sns.despine(top=True, right=True)
plt.show()
```

**En el grupo de jóvenes podemos ver graficamente las 4 modas existentes (con 4 registros cada una), en adultos se ve solo una moda (con 10 registros). Los niveles de glucosa basal son más altos y concentrados en el grupo de adultos**. 

```python
# Calculo de CV para 'G1'
g1_cv = round((g1_std / g1_mean) * 100, 2)

# Calculo de CV para 'G2'
g2_cv = round((g2_std / g2_mean)* 100, 2)

print('Coeficientes de Variación')
print(f"Grupo Jóvenes (g1): {g1_cv}")
print(f"Grupo Adultos (g2): {g2_cv}")
```

#### Resultados:      
**La media SI es representativa para Grupo 1**.       
**La media SI es representativa para Grupo 2**.      


Aunque el **grupo de jóvenes (g1) presenta distribución multimodal** (75, 79, 82, 90), con cuatro subgrupos lo que no sorprende, porque los cuerpos siguen en desarrollo, **la media sigue siendo representativa** ya que no se aleja significativamente de las modas y el CV es inferior al 15%.

El **grupo de adultos (g2)** muestra una **media aún más representativa** (CV = 8.22%), con distribución unimodal (moda en 88) y valores cercanos entre media (89.40) y moda.

Los adultos presentan niveles de glucosa basal más homogéneos, más altos y con una media más representativa que los jóvenes.

#### Shapiro-Wilk (Prueba de normalidad)
Tenemos para el grupo de jóvenes 35 muestras y 30 para el de adultos. Al tener pocos datos es conveniente hacer la prueba de Shapiro-Wilk.

### Grupo de Jóvenes (grupo 1)

#### Estudio de la reación de glucosa basal y glucosa después de la ingesta (60 min)

### a)	Relación lineal existente entre glucosa basal y glucosa a 60 min. 

```python
# Gráfico de relación
# Scatter plot G1
g1_before
g1_after
plt.figure(figsize=(6, 4))
plt.scatter(g1_before, g1_after, color='#366FE5', alpha=0.8, s=80)

#  Línea
m, b = np.polyfit(g1_before, g1_after, 1)  # m=pendiente b=intercepto
x_line = np.array([min(g1_before), max(g1_before)])
y_line = m * x_line + b
sns.regplot(x=g1_before, y=g1_after, scatter=False, color="black")

plt.plot(x_line, y_line, color='gray', linewidth=3, label=f'y = {m:.2f}x + {b:.2f}')
plt.xlabel('Glucosa Basal(mg/dL) ', fontsize=12)
plt.ylabel('Glucosa a 60 min(mg/dL) ', fontsize=12)
plt.title('Relación entre Glucosa Basal y Glucosa a 60 min en Grupo 1', fontsize=14, fontweight='bold', pad=12)

plt.grid(True, alpha=0.3)
sns.despine(top=True, right=True)
plt.show()
```

Los puntos muestran un poco de dispersión alrededor de la línea de tendencia, pero no se alejan y siguen un patrón lineal ascendente. La banda de confianza (área sombreada gris) es relativamente estrecha, lo que sugiere que el modelo lineal tiene un ajuste razonable.

Existen dos valores muy marcados que no siguen la línea, uno en  ~65, 131 en glucosa basal muy baja y otro en ~102, 172 con un respuesta muy elevada, siendo ambos valores atípicos. Los datos estan en un rango para glucosa basal de aproximadamente 65-102 mg/dL y  glucosa a 60 min de aproximadamente 131-172 mg/dL .

Podemos ver como a mayor nivel de glucosa basal, será mayor el nivel de glucosa después de la ingesta (a 60 min). La **relación es positiva** entre ambas variables. Recordemos que en grupo 1 es de jóvenes, con cuerpos todavía en desarrollo, por que que es de esperarse econtrar esta variabilidad en los datos. 

```python
# Correlación de Pearson y P-value  entre glucosa basal y a 60 min
g1_r, p_value = ss.pearsonr(g1_before, g1_after)

alpha = 0.05

print(f"P-Value: {p_value:.8f}")
print(f"Coeficiente r de Pearson: {g1_r:.3f}")

if p_value < alpha:
    print("Existe una correlación significativa entre la glucosa basal y la glucosa a 60 min.")
else:
    print("No existe una correlación significativa entre la glucosa basal y la glucosa a 60 min.")
```

Después de obtener el **coeficiente de correlación de Pearson de r = 0.796** podemos confirmar que existe una relación lineal positiva entre las dos medidas de glucosa. Con el **P-valor** en practicamente **en 0**, confirmamos que los resultados no son casualidad. **Los jóvenes con niveles altos de glucosa basal al inicio suelen tener un nivel mayor después de la ingesta**. 

### b) Modelo de regresión lineal 

#####  Modelo que explica el nivel de glucosa en sangre a los 60 minutos en función del nivel basal del paciente. 
##### Estimación para un paciente cuyo nivel basal es 83 mg/dL .


```python
# Modelo de Regresión lineal

# Definir variables
X =  g1_before.values.reshape(-1, 1) # glucosa basal
y =  g1_after # glucosa a 60 min

# Modelo
modelo = LinearRegression()

# Entrenamiento de modelo
modelo.fit(X, y)

# Dato basal nuevo
predicción = modelo.predict([[83]]) [0]
print(f"Predicción aprox. de glucosa a 60 min: {predicción:.2f}(mg/dL )")

# Cálculo de pendiente e intercepto
pendiente = float(modelo.coef_.item())  
intercepto = float(modelo.intercept_.item())
print(f"Modelo de Regresión lineal: y = {pendiente:.3f} * X + {intercepto:.3f}")
```

En la pendiente: por cada 1 mg/dL de glucosa basal, la glucosa a 60 minutos aumenta 0.697  mg/dL .    
En el intercepto: si la glucosa basal esta en cero, el valor a 60 minutos será de 91.384 mg/dL .

Entonces, si tenemos un paciente con glucosa basal de 83 el cálculo será: 
$$ \text{predicción} = 0.697 \times 83 + 91.384 \approx 149.2 $$

Lo que corresponde a la predicción generada por el modelo, **el nivel de glucosa basal a 83 mg/dL será de 149.23 mg/dL después de 60 min**.

```python
# Calidad del Modelo de Regresión Lineal
# R-cuadrado (R²)
r2 = modelo.score(X, y)  
print(f"R² = {r2:.3f}")  

# Error cuadrático medio (MSE) y raíz (RMSE)
mse = mean_squared_error(y, modelo.predict(X))
rmse = np.sqrt(mse)
print(f"RMSE = {rmse:.2f}")
```

Según el R² el modelo registra 63.4% de la variabilidad total  correctamente, o sea, que la variacón de la glucosa a 60 min se explica debido a la glucosa basal, el resto puede deberse a lo mencionado anteriormente, que el cuerpo sigue en desarrollo y no llega a niveles más homogéneos como en los de un cuerpo adulto. 
El RMSE con 4.61 indica que puede equivocarse esa cantidas en mg/dL por predicción. 

### c)	¿Qué porcentaje del nivel de glucosa en sangre pasados 60 minutos queda no queda explicado por el anterior modelo?

# PARTE III
## Intervalos de Confianza 

### a) ¿Es 88 mg/dL el nivel medio de glucosa en sangre en el momento de la ingestión en los jóvenes?

##### Exploración de intervalo de confianza al 95% y al 99%.    
##### Contraste de hipótesis.

#### Estimación del intervalo de confianza (95% y 99%) para la media de glucosa en jóvenes.

```python
# Estadística descriptiva
n = 35
g1_before_mean = g1['basal_glucose'].mean()
print(f"Media: {round(g1_before_mean,2)}") 
g1_before_sd = g1['basal_glucose'].std()
print(f"Desviación estándar: {round(g1_before_sd,2)}") 
grados_libertad = n - 1 # 35-1 (tamaño muestral - 1)
print(f"Tamaño muestral: {n} \nGrados de libertad: {grados_libertad}")

# Calcular error estándar
se = g1_before_sd/np.sqrt(n)
print(f"Error estándar: {round(se,2)}") 

t_95 = ss.t.ppf(0.975, grados_libertad)  # 2.032
t_99 = ss.t.ppf(0.995, grados_libertad)  # 2.728

# Líminte inferior
inf_ic_95 = g1_before_mean - t_95 * se
sup_ic_95 = g1_before_mean + t_95 * se
# Líminte superir 
inf_ic_99  = g1_before_mean - t_99 * se
sup_ic_99  = g1_before_mean + t_99 * se
print('\nLímites')
print(f"IC 95%: [{inf_ic_95:.2f}, {sup_ic_95:.2f}]")
print(f"IC 99%: [{inf_ic_99:.2f}, {sup_ic_99:.2f}]")
```

#### IC 95%: El nivel medio de glucosa basal en jóvenes ES DIFERENTE a 88 mg/dL .

#### IC 99%: El nivel medio de glucosa basal en jóvenes PODRÍA SER 88 mg/dL .


#### Contraste de hipótesis para la media de glucosa en jóvenes

```python
# Contraste de hipótesis para la media de glucosa en jóvenes
media = 88

t_stat = (g1_before_mean  - media) / se
print(f"\nt statistic: {t_stat:.2f}")

# p-value
p_value = 2 * (1 - ss.t.cdf(abs(t_stat), grados_libertad))
print(f"p-value: {p_value:.3f}")

# IC 95
a_95 = 0.05
if p_value < a_95:
    print("\nIC 95%: El nivel medio de glucosa basal en jóvenes es diferente a 88 mg/dL .")
else:
    print("\nIC 95%: El nivel medio de glucosa basal en jóvenes podría ser 88 mg/dL .")

# IC 99
a_99 = 0.01
if p_value < a_99:
    print("\nIC 99%: El nivel medio de glucosa basal en jóvenes es diferente a 88 mg/dL .")
else:
    print("\nIC 99%: El nivel medio de glucosa basal en jóvenes podría ser 88 mg/dL .")
```

### b)	Obtener los intervalos de confianza al 95%  para la diferencia de medias en el nivel basal de glucosa entre adultos y jovenes e interpreta los resultados. ¿Se puede concluir que el nivel basal de glucosa de los jóvenes y los adultos es el mismo con nivel de significación del 5%? .Suponiendo que se cumplen las condiciones iniciales teóricas para obtener los intervalos de confianza

##### Diferencia de medias

```python
# Estadística descriptiva para glucosa basal en adultos
g2_before_mean = g2['basal_glucose'].mean()
print(f"Media: {round(g2_before_mean,2)}") 
g2_before_sd = g2['basal_glucose'].std()
print(f"Desviación estándar: {round(g2_before_sd,2)}") 
grados_libertad_a = 29 # 30-1 (tamaño muestral - 1


# Jóvenes (grupo 1)
n1 = 35
g1_before_mean = 84.69
g1_before_sd = 8.84

# Adultos (grupo 2)
n2 = 30 
g2_before_mean = 89.40
g2_before_sd = 7.35

# Diferencia de medias (Adultos - Jóvenes)
dif_mean = g2_before_mean - g1_before_mean 
print(f"Diferencia de medias (Adultos - Jóvenes): {dif_mean:.2f} mg/dL ")

# Error estándar de la diferencia (método de Welch)
se_2 = np.sqrt((g1_before_sd**2/n1) + (g2_before_sd**2/n2))
print(f"Error estándar de la diferencia: {se_2:.4f}")

# Grados de libertad aproximados (Welch-Satterthwaite)
num = (g1_before_sd**2/n1 + g2_before_sd**2/n2)**2
den = ((g1_before_sd**2/n1)**2)/(n1-1) + ((g2_before_sd**2/n2)**2)/(n2-1)
grados_libertad_2 = num / den
print(f"Grados de libertad aproximados: {grados_libertad_2:.1f}")

# Valor t crítico para IC 95% 
t_critico = ss.t.ppf(0.975, grados_libertad_2)
print(f"Valor t crítico (95%): {t_critico:.3f}")

# Margen de error
margen = t_critico * se_2
print(f"Margen de error: {margen:.3f}")

# Intervalo de confianza
ic_inf = dif_mean - margen
ic_sup = dif_mean + margen
print(f"IC 95% para diferencia (Adultos - Jóvenes): [{ic_inf:.2f}, {ic_sup:.2f}] mg/dL ")

# Prueba de hipótesis
t_estadistico = dif_mean / se_2
p_valor = 2 * (1 - ss.t.cdf(abs(t_estadistico), grados_libertad_2))
print(f"\nTest de hipótesis:")
print(f"Estadístico t: {t_estadistico:.3f}")
print(f"p-valor: {p_valor:.4f}")

# Interpretación CORREGIDA
if ic_inf > 0:  # Todo el intervalo está por encima de 0
    print(f"\nExiste diferencia significativa: Adultos tienen entre {ic_inf:.1f} y {ic_sup:.1f} mg/dL MÁS que jóvenes.")
elif ic_sup < 0:  # Todo el intervalo está por debajo de 0
    print(f"\nExiste diferencia significativa: Adultos tienen entre {abs(ic_sup):.1f} y {abs(ic_inf):.1f} mg/dL MENOS que jóvenes.")
else:  # El intervalo incluye el 0
    print("\nNo existe diferencia significativa entre las medias (el IC incluye el 0).")
```

#### Las medias NO son iguales en ambos grupos, los adultos tienen entre 0.7 y 8.7 mg/dL más que los jóvenes, claro está que no es igual a 0. La prueba t de Welch da un p-value de 0.022 (< 0.05), por lo que se rechaza la hipótesis nula de igualdad de medias.

### c)	Se quiere estudiar la proporción de la población con un nivel basal de glucosa superior a 95 mg/dL (prediabetes). A partir de la muestra del fichero (tomando todos los datos) obtener un intervalo de confianza al 98% y contrastar la hipótesis que la proporción de la población con glucosa superior a 95 mg/dL es 0,15 con nivel de significación del 5%.


**H₀ (Hipótesis nula):**     
La proporción de la población con glucosa basal superior a 95 mg/dL es igual al 15 % (0,15).

```python
df['prediabetes'] = (df['basal_glucose'] > 95).astype(int)
total_prediabetes = df.loc[df['prediabetes'] == 1, 'prediabetes'].sum()
print(f"Total de pacientes prediabeticos (>95 mg/dL ): {total_prediabetes}")
```

```python
# Tamaño de la muestra
n_p = 65
# p-hat
proporcion_prediabetes  = total_prediabetes / n_p 
print(f"Porcentaje de pacientes con prediabetes: {proporcion_prediabetes:.4f}")
```

```python
# Error estándar
se_p = np.sqrt(proporcion_prediabetes * (1 - proporcion_prediabetes) / n_p)
print(f"Error estándar (SE): {se_p:.6f}")

# Z para IC 98%
p_ic_98 = norm.ppf(0.99) 
print(f"Valor z crítico (98%): {p_ic_98:.6f}")

# Margen de error para IC 98%
margen_error_98 = p_ic_98 * se_p
print(f"Margen de error (IC 98%): {margen_error_98:.6f}")

# Intervalos de confianza
p_inf_ic_98 = proporcion_prediabetes - p_ic_98  * se_p
p_sup_ic_98 = proporcion_prediabetes + p_ic_98  * se_p
print(f"IC 98%:[{p_inf_ic_98:.6f}, {p_sup_ic_98:.6f}]")


# Prueda de hipótesis (H0 ->  p = 0.15)
p0 = 0.15
z_stat = (proporcion_prediabetes - p0) / np.sqrt(p0*(1-p0)/n_p)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))
print(f"\nTest de hipótesis:")
print(f"z-value: {z_stat:.6f}")
print(f"p-value: {p_value:.6f}")

if p_value > 0.05:
    print("\nNo se rechaza H₀ (p = 0.15). No hay evidencia suficiente para afirmar que la proporción difiera del 15%.")
else:
    print("\nSe rechaza H₀ (p = 0.15). Existe evidencia para concluir que la proporción es diferente del 15%.")
```

Esto se cofirma al ver que  el 15% está dentro del IC 98% 6.1% - 27.7% y el p-valor de 0.6641 > 0.05, entonces no se rechaza H₀.
